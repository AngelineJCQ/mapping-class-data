```python
import geopandas as gpd
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
counties = gpd.read_file('countyTypologyCodes.geojson')
```

```python
counties.head()
```

```python
counties.Economic_Type_Label.value_counts()
```

```python
counties['sum'] = counties['Low_Education_2015_Update'] + \
                  counties['Low_Employment_Cnty_2008_2012_25_64'] + \
                  counties['Pop_Loss_2010'] + \
                  counties['Retirement_Dest_2015_Update'] + \
                  counties['Persistent_Poverty_2013'] + \
                  counties['Persistent_Related_Child_Poverty_2013']
```

```python
import numpy as np
```

```python
counties = counties.rename(columns={'Low_Education_2015_Update': "Low Education",'Low_Employment_Cnty_2008_2012_25_64': 'Low Employment','Pop_Loss_2010': 'Population Loss', \
                 'Retirement_Dest_2015_Update': 'Retirement Destination', \
                 'Persistent_Poverty_2013': 'Persistent Poverty', \
                 'Persistent_Related_Child_Poverty_2013': 'Persistent Child Poverty'})
```

```python
headers = ["Low Education",'Low Employment','Population Loss','Retirement Destination','Persistent Poverty','Persistent Child Poverty']
for col in headers:
    counties[col] = counties[col].map({1:col, 0:np.nan}) 
```

```python
counties.fillna('', inplace=True)
```

```python
counties.head()
```

```python
counties.to_file('countiesTypologyCodesNew.geojson', driver='GeoJSON')
```
